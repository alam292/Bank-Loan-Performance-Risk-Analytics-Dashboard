### Insights

```python
plt.figure(figsize = (10,5))
plotnumber = 1
for i in df_cont:
    plt.subplot(1,3,plotnumber)
    sns.histplot(x = df_cont[i] , kde = True)
    plotnumber += 1
plt.tight_layout()
```

### Bivariate Analysis

```python
plt.figure(figsize = (10,8))
plotnumber = 1
for i in df_cat:
    plt.subplot(3,3,plotnumber)
    sns.countplot(x = df[i] , hue = df.Loan_Status)
    plt.xticks(rotation = 45)
    plotnumber += 1
plt.tight_layout()
```

### Insights

```python
plt.figure(figsize = (10,8))
plotnumber = 1
for i in df_cont:
    plt.subplot(3,3,plotnumber)
    sns.histplot(x = df_cont[i] ,kde = True, hue = df.Loan_Status)
    plotnumber += 1
plt.tight_layout()
```

## Data Cleaning

### Handling with null values

```python
df.isnull().sum()
```

```python
df.loc[df.Gender.isnull()]
```

```python
df.Gender.value_counts()
```

```python
df.loc[df.Gender.isnull() , 'Gender']
```

```python
df.loc[df.Gender.isnull() , 'Gender'] = 'Male'
```

```python
df.Gender.value_counts()
```

```python
df.Gender.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.Married.value_counts()
```

```python
df.loc[df.Married.isnull() , 'Married'] = 'Yes'
```

```python
df.Married.value_counts()
```

```python
df.Dependents.value_counts()
```

```python
df.loc[df.Dependents.isnull() , 'Dependents'] = '0'
```

```python
df.Dependents.value_counts()
```

```python
df.Self_Employed.value_counts()
```

```python
df.loc[df.Self_Employed.isnull() , 'Self_Employed'] = 'No'
```

```python
df.Self_Employed.value_counts()
```

```python
df.isnull().sum()
```

```python
df.loc[df.LoanAmount.isnull(), 'LoanAmount']
```

```python
len(df.LoanAmount)
```

```python
len(df.LoanAmount.dropna())      # remove missing value
```

```python
np.median(df.LoanAmount.dropna())
```

```python
df.loc[df.LoanAmount.isnull(), 'LoanAmount'] = np.median(df.LoanAmount.dropna())
```

```python
df.head()
```

```python
df.Loan_Amount_Term.value_counts()
```

```python
df.loc[df.Loan_Amount_Term.isnull() , 'Loan_Amount_Term'] =360.0
```

```python
df.isnull().sum()
```

```python
df.Credit_History.value_counts()
```

```python
df.loc[df.Credit_History.isnull() ,'Credit_History'] = 1.0
```
